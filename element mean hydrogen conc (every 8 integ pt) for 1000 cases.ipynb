# find max

```python
beta=np.loadtxt('beta_element_mean.txt')
beta_max_index=[]
for i in range(0,1000):
    beta_index=np.where(beta[:,i] == np.amax(beta[:,i]))
    beta_max_index.append(beta_index[0]+1)
np.savetxt('beta_max_index_element_wise.txt',beta_max_index)
```
