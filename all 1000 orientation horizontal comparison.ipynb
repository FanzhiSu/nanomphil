```python
import numpy as np  
import matplotlib.pyplot as plt
zero=np.zeros((1000,3))
data_max_beta= np.loadtxt('beta_max.txt')
data_max_lam1= np.loadtxt('lam1_max.txt')
data_max_lam2= np.loadtxt('lam2_max.txt')
data_max_lam3= np.loadtxt('lam3_max.txt')
data=np.vstack((data_max_beta,data_max_lam1))
data=np.vstack((data,data_max_lam2))
data=np.vstack((data,data_max_lam3))

data=data.T

data
zero[:,2]=list(range(1,1001))
```

```python
for i in range(0,1000):
    index=np.where(data[i,:]== np.amax(data[i,:]))
    zero[i,0]=data[i,index[0]]
    zero[i,1]=index[0]

import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
print(zero)
```

```python
x_beta=[]
y_beta=[]
x_lam1=[]
y_lam1=[]
x_lam2=[]
y_lam2=[]
x_lam3=[]
y_lam3=[]
for i in range(0,1000):
    if zero[i,1]==0:
        x_beta.append(zero[i,2])
        y_beta.append(zero[i,0])
    elif zero[i,1]==1:
        x_lam1.append(zero[i,2])
        y_lam1.append(zero[i,0])
    elif zero[i,1]==2:
        x_lam2.append(zero[i,2])
        y_lam2.append(zero[i,0])
    else:
        x_lam3.append(zero[i,2])
        y_lam3.append(zero[i,0])
```
