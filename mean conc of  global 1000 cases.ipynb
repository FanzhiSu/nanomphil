```python
import numpy as np,matplotlib.pyplot as plt
plt.figure(figsize=[10,10])

data1 = np.loadtxt("beta_centre_c.txt")
data2 = np.loadtxt("lam1_c.txt")
data3 = np.loadtxt("lam2_c.txt")
data4 = np.loadtxt("lam3_c.txt")

def ecdf_(data):
	"""Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
	n = len(data)
    # x-data for the ECDF: x
	x = np.sort(data)
    # y-data for the ECDF: y
	y = np.arange(1,n+1)/n
	return x, y


# Compute ECDFs
x_1,y_1 = ecdf_(data1/np.linalg.norm(data1))
x_2,y_2 = ecdf_(data2/np.linalg.norm(data2))
x_3,y_3 = ecdf_(data3/np.linalg.norm(data3))
x_4,y_4 = ecdf_(data4/np.linalg.norm(data4))
x_5,y_5 = ecdf_(data1)
x_6,y_6 = ecdf_(data2)
x_7,y_7 = ecdf_(data3)
x_8,y_8 = ecdf_(data4)





percentiles = np.array([2.5,25,50,75,97.5])
ptiles_data1 = np.percentile(x_1,percentiles)
ptiles_data2 = np.percentile(x_2,percentiles)
ptiles_data3 = np.percentile(x_3,percentiles)
ptiles_data4 = np.percentile(x_4,percentiles)


ptiles_data5 = np.percentile(x_5,percentiles)
ptiles_data6 = np.percentile(x_6,percentiles)
ptiles_data7 = np.percentile(x_7,percentiles)
ptiles_data8 = np.percentile(x_8,percentiles)



###########""bootstrapping
# def bootstrap(data,num_bootsraps):
# 	for i in range(num_bootsraps):
#     	# Generate bootstrap sample: bs_sample
# 		bs_sample = np.random.choice(data,size=len(data))

# 	    # Compute and plot ECDF from bootstrap sample
# 		x, y = ecdf_(bs_sample)
# 		_ = plt.plot(x, y, marker='.', linestyle='none',color='gray', alpha=0.05)
####done



#bootstrap(data1,500)



# Plot all ECDFs on the same plot
plt.plot(x_1,y_1,marker='.',linestyle="None",linewidth="9",c="g",label="cubic beta ")
plt.plot(x_2,y_2,marker='.',linestyle="None",linewidth="9",c="b",label="cuboidal beta")
plt.plot(x_3,y_3,marker='.',linestyle="None",linewidth="9",c="r",label="2D cross-shaped beta")
plt.plot(x_4,y_4,marker='.',linestyle="None",linewidth="9",c="m",label="3D cross-shaped beta")
# Make nice margins
plt.margins(0.02)

# Annotate the plot
plt.xlabel('normalised hydrogen concentration / ppm',fontsize=20)
_ = plt.ylabel('ECDF',fontsize=20)

plt.title('CDF',fontsize=20)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 
# Display the plot

plt.legend(loc='upper left',prop={'size': 15})
plt.show()

plt.figure(figsize=[10,10])
#_ = plt.plot(ptiles_data, percentiles/100, marker='D', color='red',linestyle='none')#superimposing the percentile as red dotson the graphh
plt.plot(x_5,y_5,marker='.',linestyle="None",linewidth="9",c="g",label="cubic beta ")
plt.plot(x_6,y_6,marker='.',linestyle="None",linewidth="9",c="b",label="cuboidal beta")
plt.plot(x_7,y_7,marker='.',linestyle="None",linewidth="9",c="r",label="2D cross-shaped beta")
plt.plot(x_8,y_8,marker='.',linestyle="None",linewidth="9",c="m",label="3D cross-shaped beta")

# Make nice margins
plt.margins(0.02)

# Annotate the plot
plt.xlabel('hydrogen concentration / ppm',fontsize=20)
_ = plt.ylabel('ECDF',fontsize=20)

plt.title('CDF',fontsize=20)
plt.rc('xtick', labelsize=15) 
plt.rc('ytick', labelsize=15) 
# Display the plot

plt.legend(loc='upper left',prop={'size': 15})
plt.show()



```
