```python
import numpy as np
struc1=np.zeros((13824,1000))

for i in range(0,1000):  
    struc1_data=np.loadtxt('/home/clark/Desktop/mphil/send_structs/struct_1/case'+str(i)+'/c.txt')[:,1]
    struc1[:,i]=struc1_data
```

```python
struc2=np.zeros((13824,1000))
for i in range(0,1000):  
    struc2_data=np.loadtxt('/home/clark/Desktop/mphil/send_structs/struct_2/case'+str(i)+'/c.txt')[:,1]
    struc2[:,i]=struc2_data
```

```python
struc3=np.zeros((13824,1000))
for i in range(0,1000):  
    struc3_data=np.loadtxt('/home/clark/Desktop/mphil/send_structs/struct_3/case'+str(i)+'/c.txt')[:,1]
    struc3[:,i]=struc3_data
```

```python
struc1_average=np.zeros((1,1000))
for j in range(0,13824,8):
    struc1_average=np.vstack((struc1_average,np.divide(np.sum(struc1[j:j+8,:],axis=0),8)))
struc1_average=struc1_average[1:,:]
```

```python
struc2_average=np.zeros((1,1000))
for j in range(0,13824,8):
    struc2_average=np.vstack((struc2_average,np.divide(np.sum(struc2[j:j+8,:],axis=0),8)))
struc2_average=struc2_average[1:,:]
```

```python
struc3_average=np.zeros((1,1000))
for j in range(0,13824,8):
    struc3_average=np.vstack((struc3_average,np.divide(np.sum(struc3[j:j+8,:],axis=0),8)))
struc3_average=struc3_average[1:,:]
```

```python
np.savetxt('struc1_element_mean.txt',struc1_average)
np.savetxt('struc2_element_mean.txt',struc2_average)
np.savetxt('struc3_element_mean.txt',struc3_average)
```

# find max

```python
struc1=np.loadtxt('struc1_element_mean.txt')
struc1_max_index=[]
for i in range(0,1000):
    struc1_index=np.where(struc1[:,i] == np.amax(struc1[:,i]))
    struc1_max_index.append(struc1_index[0]+1)
np.savetxt('struc1_max_index_element_wise.txt',struc1_max_index)
```

```python
lam1=np.loadtxt('struc2_element_mean.txt')
lam1_max_index=[]
for i in range(0,1000):
    lam1_index=np.where(lam1[:,i] == np.amax(lam1[:,i]))
    lam1_max_index.append(lam1_index[0]+1)
np.savetxt('struc2_max_index_element_wise.txt',lam1_max_index)
```

```python
lam2=np.loadtxt('struc3_element_mean.txt')
lam2_max_index=[]
for i in range(0,1000):
    lam2_index=np.where(lam2[:,i] == np.amax(lam2[:,i]))
    lam2_max_index.append(lam2_index[0]+1)
np.savetxt('struc3_max_index_element_wise.txt',lam2_max_index)
```
